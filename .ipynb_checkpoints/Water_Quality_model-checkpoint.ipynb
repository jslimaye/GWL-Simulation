```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("wq_final.csv")
```

```python
data['Month'] = pd.to_datetime(data['Month'])
```

```python
data.set_index('Month',inplace=True)
```

```python
data['year'] = data.index.year  # Extract year
for col in data.columns:
    # data[col] = data[col].astype(str).str.extract(r'([\d.]+)')
    data[col].apply(pd.to_numeric,errors='coerce')
    data[col] = data.groupby('year')[col].transform(lambda x: x.fillna(x.mean()))

# Drop helper column
data.drop(columns=['year'], inplace=True)
```

```python
data.isna().sum()
```

```python
data_scale=dict()
for col in data.columns:
    data_scale[col] = [data[col].min(),data[col].max()]
    data[col] = (data[col] - data[col].min())/(data[col].max() - data[col].min())
    
```

```python
data_scale
```

```python
data
```

```python
import numpy as np

def prepare_data(data, look_back):
    data = np.array(data)  # Ensure data is a NumPy array
    X, Y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back), :])  # Selecting 'look_back' timesteps for all columns
        Y.append(data[i + look_back, :])  # Predicting the next timestep for all columns
    return np.array(X), np.array(Y)
```

```python
look_back = 24
X,y = prepare_data(data,look_back)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(64,return_sequences=True, input_shape=(look_back,7)))
model.add(Dense(64))
model.add(Dense(128))
model.add(LSTM(128))
model.add(Dense(128))
model.add(Dense(64))
model.add(Dense(32))
model.add(Dense(7,activation='relu'))

model.compile(optimizer='adam', loss='mean_squared_error')#,metrics=['loss'])
```

```python
model.fit(X,y, batch_size=32, verbose=1, epochs=50)
```

```python
y_pred = model.predict(X)
```

```python
y_pred.shape
```

```python
y_pred[:,1]
```

```python
cols = list(data.columns)
for i in range(len(y_pred)):
    plt.plot(y_pred[:,i],color='r',label="Predicted Values")
    plt.plot(list(data[cols[i]][look_back:]),color='g', label="Actual Values")
    plt.title(f"Result of {cols[i]}")
    plt.legend()
    plt.show()
```
