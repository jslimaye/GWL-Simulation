```python
import os
```

```python
data = list()
path = "GWL_Data"
for file in os.listdir(path):
    if file.endswith(".xlsx"):
        filename = f"{path}/{file}"
        # print(filename)
        df = pd.read_excel(filename,skiprows=1)
        data.append(df)
        # print(len(df))
```

```python
def split_dataframe(df, n):
    return [df.iloc[:, i:i+n] for i in range(0, df.shape[1], n)]
```

```python
df = data[0]
merged_df = pd.DataFrame()
split_dfs = split_dataframe(df, 2)

for split_df in split_dfs:
    cols = split_df.columns
    split_df.columns = ['station',cols[0]]
from functools import reduce    
merge_df = reduce(lambda left, right: pd.merge(left, right, on='station',how='outer'), split_dfs) 
merge_df
```

```python
merge=merge_df.dropna(how='all')
merge
```

```python
for df in data:
    df.drop(columns=['Station Code'],errors='ignore')
```

```python
from functools import reduce 
from tqdm import tqdm
data_final = list()
for i in tqdm(range(len(data))):
    df = data[i]
    merged_df = pd.DataFrame()
    split_dfs = split_dataframe(df, 2)
    
    for split_df in split_dfs:
        cols = split_df.columns
        split_df.columns = ['station',cols[0]]
        # print(f"for data {i}, colums are {split_df.columns}")
    merge_df = reduce(lambda left, right: pd.merge(left, right, on='station',how='outer'), split_dfs)
    merge=merge_df.dropna(how='all')
    data_final.append(merge)

print(len(data))
print(len(data_final))
```

```python
def clean_value(val):
    if val==None:
        return val
    if type(val)==str:
        val = val.strip()  # Remove leading/trailing spaces
        parts = val.split(" - ")  # Split by ' - '
        
        # If the format is "X - X", return only "X", else keep original
        return parts[0] if len(parts) == 2 and parts[0] == parts[1] else val
```

```python
data_cleaned = list()
for df in data_final:
# df = data_final[1]
# df.info()
    for col in df.columns:
        if df[col].dtype=='object':
            df[col] = df[col].map(clean_value)
    data_cleaned.append(df)
```

```python
print(data_cleaned[1])
```

```python
gwl_data = reduce(lambda left, right: pd.merge(left, right, on='station',how='outer'), data_cleaned)
```

```python
gwl_data.isna().sum()
```

```python
gwl_data
```

```python
gwl_final = gwl_data.T
gwl_final
```

```python
gwl_final.to_csv("GWL_final.csv",header=False)
```

# Water Quality Data

```python
cols = ['Month','pH','dissolved_oxygen','BOD','COD','Nitrate','Fecal_Coliform','WQI']
```

```python
data = dict()
path = "WQ_Data"
for file in os.listdir(path):
    if file.endswith(".xlsx"):
        filename = f"{path}/{file}"
        # print(filename)
        df = pd.read_excel(filename,skiprows=3)
        data[str(file.split()[0])] = df
        print(len(df))
```

```python
data.keys()
```

```python
for key in data.keys():
    data[key] = data[key].iloc[:-3].dropna(axis=1,how="all")
```

```python
data['2021']
```

```python
for key in data.keys():
    data[key].columns = cols
```

```python
data['2020']
```

```python
for key in data.keys():
    month_list = list(data[key]['Month'])
    for i in range(len(month_list)):
        month_list[i] = month_list[i]+"-"+key
    # print(month_list)
    data[key]['Month'] = month_list
```

```python
wq_final = pd.DataFrame()
for key in data.keys():
    wq_final = pd.concat([wq_final, data[key]],ignore_index=True)
wq_final
```

```python
wq_final.to_csv("wq_final.csv",index=False)
```

# Rainfall Data

```python
rain_data = pd.read_csv("Rainfall data_2000-2023_18.52_73.85 (1).csv")
```

```python
rain_data
```

```python
rain_data.info()
```

```python
rain_data['DateTime'] = pd.to_datetime(rain_data['DateTime'],format="%d-%m-%Y")
```

```python
rain_data.info()
```

```python
rain_final = rain_data.groupby(rain_data['DateTime'].dt.to_period('M')).mean()
rain_final
```

```python
rain_final.columns = ['Month','Rainfall']
```
